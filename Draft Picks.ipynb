```python
import pandas as pd
draft = pd.read_csv('nfl_draft.csv', low_memory = False)
draft #verify data is here
```

Sum up draft scores:

```python
years = set(draft['Year'])
data = list()
for year in years:
    year_draft = draft[draft['Year'] == year] #split by year
    teams = set(year_draft['Tm'])
    t_dic = dict.fromkeys(teams, 0)
    for team in teams:
        t_draft = year_draft[year_draft['Tm'] == team]
        t_dic[team] = t_draft['DrAV'].sum()
    data.append([year, t_dic])

year_list = list()
team_list = list()
score_list = list()
for x in data:
    # go year by year
    for team, score in x[1].items():
        #get every score
        year_list.append(x[0])
        team_list.append(team)
        score_list.append(score)  
```

```python
from bokeh.io import output_notebook, show
output_notebook()
```

```python
from bokeh.charts import Bar
from bokeh.charts.attributes import cat, color
from bokeh.charts.operations import blend
from bokeh.models import HoverTool, ColumnDataSource


df = pd.DataFrame({
        'year':year_list,
        'team':team_list,
        'score':score_list
    })

source = ColumnDataSource(
        data=df
    )

hover = HoverTool(
        tooltips=[
            ("Team", "@team"),
            ("Draft Score", "@height")
        ]
    )
p = Bar(df, label='year', values='score', stack='team',
        title="Yearly Draft Score By Team", tools=[hover], legend=False, source=source, color=color(columns='team', 
                                                                                            palette=["#97233F", "#C9233F", "#2A0365", "#0F4589", "#0088D4", "#DF6108", "#F03A16", "#322820",
  "#0D254C", "#DF6108", "#006DB0", "#313F36", "#B20032", "#163F83", "#007198", "#B20032",
 "#005E6A", "#3B0160", "#0D254C", "#C9B074", "#192E6C", "#313F36", "#C4C8CB", "#003B48",
 "#F2C800", "#0B264D", "#54BA4C", "#840026", "#0D254C", "#B20032", "#4C96C4", "#7A2D39"], 
                                                                                            sort=False))

show(p)
```

```python
#Normalized Data 
score_list_n = list()
for x in data:
    #go year by year
    year_score = list()
    for team, score in x[1].items():
        year_score.append(score)
    s = sum(year_score)
    normalized = [100*float(i)/s for i in year_score]
    score_list_n.extend(normalized)
```

```python
df_1 = pd.DataFrame({
        'year':year_list,
        'team':team_list,
        'score':score_list_n
    })

source = ColumnDataSource(
        data=df_1
    )

hover = HoverTool(
        tooltips=[
            ("Team", "@team"),
            ("Draft Score", "@height")
        ]
    )
p2 = Bar(df_1, label='year', values='score', stack='team',
        title="Yearly Normalized Draft Score By Team", tools=[hover], legend=False, source=source, color=color(columns='team', 
                                                                                            palette=["#97233F", "#C9233F", "#2A0365", "#0F4589", "#0088D4", "#DF6108", "#F03A16", "#322820",
  "#0D254C", "#DF6108", "#006DB0", "#313F36", "#B20032", "#163F83", "#007198", "#B20032",
 "#005E6A", "#3B0160", "#0D254C", "#C9B074", "#192E6C", "#313F36", "#C4C8CB", "#003B48",
 "#F2C800", "#0B264D", "#54BA4C", "#840026", "#0D254C", "#B20032", "#4C96C4", "#7A2D39"], 
                                                                                            sort=False))

show(p2)
```

```python
years = set(draft['Year'])
data = list()
for year in years:
    year_draft = draft[draft['Year'] == year] #split by year
    roun = set(year_draft['Rnd'])
    r_dic = dict.fromkeys(roun, 0)
    for r in roun:
        r_draft = year_draft[year_draft['Rnd'] == r]
        r_dic[r] = r_draft['DrAV'].sum()
    data.append([year, r_dic])
year_list = list()
round_list = list()
score_list = list()
for x in data:
    # go year by year
    for roun, score in x[1].items():
        #get every score
        year_list.append(x[0])
        round_list.append(roun)
        score_list.append(score)  
#Normalized Data 
score_list_n = list()
for x in data:
    #go year by year
    year_score = list()
    for team, score in x[1].items():
        year_score.append(score)
    s = sum(year_score)
    normalized = [100*float(i)/s for i in year_score]
    score_list_n.extend(normalized)
```

```python
import bokeh.palettes as bp
df_2 = pd.DataFrame({
        'year':year_list,
        'round':round_list,
        'score':score_list_n
    })

source = ColumnDataSource(
        data=df_2
    )

hover = HoverTool(
        tooltips=[
            ("Round", "@round"),
            ("Draft Score", "@height")
        ]
    )
p3 = Bar(df_2, label='year', values='score', stack='round',
        title="Yearly Normalized Draft Score By Round", tools=[hover], legend=False, source=source, color=color(columns='round', 
                                                                                            palette=bp.Paired12, 
                                                                                            sort=False))

show(p3)
```
